```python
import folium
from folium.plugins import MarkerCluster
import pandas as pd

# Load the SpaceX launch data
spacex_df = pd.read_csv('spacex_launch_geo.csv')

# Create a map centered on the mean latitude and longitude of the launch sites
center_lat = spacex_df['Lat'].mean()
center_lon = spacex_df['Long'].mean()
site_map = folium.Map(location=[center_lat, center_lon], zoom_start=5)

# Create a MarkerCluster object
marker_cluster = MarkerCluster().add_to(site_map)

# Function to assign color to launch outcome
def assign_marker_color(launch_outcome):
    if launch_outcome == 1:
        return 'green'
    else:
        return 'red'

spacex_df['marker_color'] = spacex_df['class'].apply(assign_marker_color)

# Iterate through the dataframe and add markers to the map
for idx, row in spacex_df.iterrows():
    folium.Marker(
        location=[row['Lat'], row['Long']],
        popup=f"Launch Site: {row['Launch Site']}<br>Success: {'Yes' if row['class'] == 1 else 'No'}",
        icon=folium.Icon(color=row['marker_color'], icon='info-sign')
    ).add_to(marker_cluster)

# Display the map
site_map
```

Next, we need to explore and analyze the proximities of launch sites.

Now zoom in to a launch site and explore its proximity to see if you can easily find any railway, highway, coastline, etc. Move your mouse to these points and mark down their coordinates (shown on the top-left) in order to the distance to the launch site.


You can calculate the distance between two points on the map based on their `Lat` and `Long` values using the following method:


```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

```python
import folium
from folium.plugins import MousePosition
from folium.features import DivIcon
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance

# Assuming we're using the CCAFS SLC-40 launch site as an example
launch_site_lat = 28.5618571
launch_site_lon = -80.577366

# Create a map centered on the launch site
site_map = folium.Map(location=[launch_site_lat, launch_site_lon], zoom_start=10)

# Add MousePosition to get the coordinate (Lat, Long) for a mouse over on the map
formatter = "function(num) {return L.Util.formatNum(num, 5);};"
mouse_position = MousePosition(
    position='topright',
    separator=' Long: ',
    empty_string='NaN',
    lng_first=False,
    num_digits=5,
    prefix='Lat:',
    lat_formatter=formatter,
    lng_formatter=formatter,
)
site_map.add_child(mouse_position)

# Add a marker for the launch site
folium.Marker(
    [launch_site_lat, launch_site_lon],
    popup="CCAFS SLC-40 Launch Site",
    icon=folium.Icon(color='red', icon='rocket')
).add_to(site_map)

# Instructions for the user
print("Instructions:")
print("1. Run this cell to display the map.")
print("2. Find the closest point on the coastline to the launch site (red marker).")
print("3. Right-click on this coastline point to add a marker.")
print("4. The distance will be calculated and displayed on the map.")

# Add click event to the map
site_map.add_child(folium.LatLngPopup())

# Function to add marker on right-click
site_map.add_child(folium.ClickForMarker(popup="Coastline Point"))

# Display the initial map
display(site_map)

# After the user interacts with the map, prompt for the coastline coordinates
coastline_lat = float(input("Enter the latitude of the coastline point you marked: "))
coastline_lon = float(input("Enter the longitude of the coastline point you marked: "))

# Calculate the distance
distance_coastline = calculate_distance(launch_site_lat, launch_site_lon, coastline_lat, coastline_lon)

# Add a marker for the coastline point
coastline_marker = folium.Marker(
    [coastline_lat, coastline_lon],
    popup=f"Coastline Point<br>Distance: {distance_coastline:.2f} km",
    icon=folium.Icon(color='green', icon='info-sign')
)
coastline_marker.add_to(site_map)

# Add a line between the launch site and coastline point
folium.PolyLine(
    locations=[[launch_site_lat, launch_site_lon], [coastline_lat, coastline_lon]],
    color="blue",
    weight=2,
    opacity=0.8
).add_to(site_map)

# Add a label with the distance
midpoint = [(launch_site_lat + coastline_lat)/2, (launch_site_lon + coastline_lon)/2]
folium.map.Marker(
    midpoint,
    icon=DivIcon(
        icon_size=(150,36),
        icon_anchor=(75,18),
        html=f'<div style="font-size: 12pt; color : black">{distance_coastline:.2f} km</div>',
    )
).add_to(site_map)

print(f"The distance between the launch site and the closest coastline point is: {distance_coastline:.2f} Kilometers")

# Display the updated map
display(site_map)
```

```python
# Create a marker with distance to a closest city, railway, highway, etc.
# Draw a line between the marker to the launch site
import folium
from folium.features import DivIcon
from math import radians, sin, cos, sqrt, atan2

def calculate_distance(lat1, lon1, lat2, lon2):
    R = 6371  # Radius of the Earth in km
    lat1, lon1, lat2, lon2 = map(radians, [lat1, lon1, lat2, lon2])
    dlat = lat2 - lat1
    dlon = lon2 - lon1
    a = sin(dlat/2)**2 + cos(lat1) * cos(lat2) * sin(dlon/2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))
    distance = R * c
    return distance

# Launch site coordinates (example: Kennedy Space Center)
launch_site_lat, launch_site_lon = 28.5728, -80.6490

# Create a map centered on the launch site
m = folium.Map(location=[launch_site_lat, launch_site_lon], zoom_start=10)

# Add launch site marker
folium.Marker(
    [launch_site_lat, launch_site_lon],
    popup="Launch Site",
    icon=folium.Icon(color='red', icon='rocket', prefix='fa')
).add_to(m)

# Define nearby points (you would replace these with actual data)
nearby_points = {
    'City': (28.4104, -80.6090),
    'Railway': (28.5172, -80.6723),
    'Highway': (28.4056, -80.7255)
}

# Color mapping for different types of points
color_map = {'City': 'blue', 'Railway': 'green', 'Highway': 'orange'}

# Add markers and calculate distances
for point_type, coords in nearby_points.items():
    lat, lon = coords
    distance = calculate_distance(launch_site_lat, launch_site_lon, lat, lon)
    
    # Add marker
    folium.Marker(
        [lat, lon],
        popup=f"{point_type}<br>Distance: {distance:.2f} km",
        icon=folium.Icon(color=color_map[point_type], icon='info-sign')
    ).add_to(m)
    
    # Draw line
    folium.PolyLine(
        locations=[[launch_site_lat, launch_site_lon], [lat, lon]],
        color=color_map[point_type],
        weight=2,
        opacity=0.8
    ).add_to(m)
    
    # Add distance label
    folium.map.Marker(
        [(launch_site_lat + lat)/2, (launch_site_lon + lon)/2],
        icon=DivIcon(
            icon_size=(150,36),
            icon_anchor=(75,18),
            html=f'<div style="font-size: 10pt; color:{color_map[point_type]};"><b>{distance:.2f} km</b></div>'
        )
    ).add_to(m)

# Display the map
site_map
```
